```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t.isalnum() and t not in stop_words]
    return ' '.join(tokens)
```

```python
# --- 1. Extract tweet fields ---
def parse_tweet(tweet):
    try:
        if isinstance(tweet, str):
            tweet = eval(tweet)
        return {
            'tweet_text': tweet.get('text', ''),
            'screen_name': tweet.get('user', {}).get('screen_name', ''),
            'location': tweet.get('user', {}).get('location', ''),
            'description': tweet.get('user', {}).get('description', ''),
            'followers_count': tweet.get('user', {}).get('followers_count', 0),
            'friends_count': tweet.get('user', {}).get('friends_count', 0),
            'retweet_count': tweet.get('retweet_count', 0),
            'favorite_count': tweet.get('favorite_count', 0),
        }
    except Exception:
        return {f: '' for f in ['tweet_text', 'screen_name', 'location', 'description']} | \
               {f: 0 for f in ['followers_count', 'friends_count', 'retweet_count', 'favorite_count']}
```

```python
# --- 2. Extract Watson fields ---
def parse_watson(w):
    try:
        if isinstance(w, str):
            w = eval(w)
        categories = w.get('categories', [])
        cat_labels = [cat['label'] for cat in categories[:3]]
        cat_scores = [cat['score'] for cat in categories[:3]]
        return {
            'watson_language': w.get('language', ''),
            'watson_cat1': cat_labels[0] if len(cat_labels) > 0 else '',
            'watson_cat2': cat_labels[1] if len(cat_labels) > 1 else '',
            'watson_cat3': cat_labels[2] if len(cat_labels) > 2 else '',
            'watson_score1': cat_scores[0] if len(cat_scores) > 0 else 0,
            'watson_score2': cat_scores[1] if len(cat_scores) > 1 else 0,
            'watson_score3': cat_scores[2] if len(cat_scores) > 2 else 0,
        }
    except Exception:
        return {
            'watson_language': '',
            'watson_cat1': '',
            'watson_cat2': '',
            'watson_cat3': '',
            'watson_score1': 0,
            'watson_score2': 0,
            'watson_score3': 0,
        }
```

```python
# --- 3. URL feature extraction ---
def extract_url_features(url_list):
    try:
        if isinstance(url_list, str):
            url_list = eval(url_list)
        url = url_list[0] if url_list else ''
    except:
        url = ''
    parsed = urlparse(url)
    return {
        'url_domain': parsed.netloc,
        'url_https': parsed.scheme == 'https',
        'url_len': len(url),
    }
```

```python
# --- 4. Destination URL features ---
def dest_url_feats(url):
    parsed = urlparse(str(url))
    return {
        'dest_domain': parsed.netloc,
        'dest_https': parsed.scheme == 'https',
        'dest_url_len': len(str(url)),
    }
```

```python
tqdm_auto.pandas()
label_encoders = {}

def preprocess_data(df=None, mode='train'):
    print(f"[INFO] Mode: {mode}")

    if df is None:
        raise ValueError("A valid DataFrame must be provided.")

    print("[INFO] Starting preprocessing...")

    # Tweet features
    tweet_features = df['tweet'].progress_apply(parse_tweet).apply(pd.Series)
    df = pd.concat([df.drop(columns=['tweet']), tweet_features], axis=1)

    # Watson features
    watson_features = df['watson'].progress_apply(parse_watson).apply(pd.Series)
    df = pd.concat([df.drop(columns=['watson']), watson_features], axis=1)

    # URL features
    url_features = df['urls'].progress_apply(extract_url_features).apply(pd.Series)
    df = pd.concat([df.drop(columns=['urls']), url_features], axis=1)

    # Destination URL features
    dest_url_features = df['destination_url'].progress_apply(dest_url_feats).apply(pd.Series)
    df = pd.concat([df.drop(columns=['destination_url']), dest_url_features], axis=1)

    # Clean text columns
    text_cols = ['tweet_text', 'screen_name', 'location', 'description',
                 'watson_cat1', 'watson_cat2', 'watson_cat3',
                 'url_domain', 'dest_domain']
    for col in tqdm(text_cols, desc="Cleaning text columns"):
        df[f'processed_{col}'] = df[col].astype(str).progress_apply(clean_text)

    # Process text + numeric
    numeric_cols = ['followers_count', 'friends_count', 'retweet_count', 'favorite_count',
                    'watson_score1', 'watson_score2', 'watson_score3',
                    'url_https', 'url_len', 'dest_https', 'dest_url_len']

    processed_text = df[[f'processed_{col}' for col in text_cols]].agg(' '.join, axis=1)
    X_numeric = df[numeric_cols].fillna(0).astype(str).agg(' '.join, axis=1)
    X = (processed_text + ' ' + X_numeric).tolist()

    if mode == 'train':
        # Label encoding for training
        targets = ['relevant', 'type']
        Y_dict = {}
        for target in targets:
            if target in df.columns:
                le = LabelEncoder()
                Y_dict[target] = le.fit_transform(df[target].astype(str))
                label_encoders[target] = le
        return X, Y_dict.get('relevant'), Y_dict.get('type')

    elif mode == 'test':
        return X

    else:
        raise ValueError("mode must be either 'train' or 'test'")
```

```python
def eda_nlp_analysis(X_text, num_words=100, top_n_words=20):

    print("Generating NLP EDA Visualizations...")

    # Flatten all tokens from all texts
    all_tokens = [word for doc in X_text for word in word_tokenize(doc)]

    # --- 1. WordCloud ---
    word_freq = Counter(all_tokens)
    wc = WordCloud(width=800, height=400, max_words=num_words, background_color='white').generate_from_frequencies(word_freq)

    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Top {num_words} Words - WordCloud")
    plt.show()

    # --- 2. Top-N Frequent Words ---
    common_words = word_freq.most_common(top_n_words)
    words, counts = zip(*common_words)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(counts), y=list(words), palette="viridis")
    plt.title(f"Top {top_n_words} Most Frequent Words")
    plt.xlabel("Count")
    plt.ylabel("Word")
    plt.show()

    # --- 3. Document Length Histogram ---
    doc_lengths = [len(word_tokenize(doc)) for doc in X_text]
    plt.figure(figsize=(10, 5))
    sns.histplot(doc_lengths, bins=20, kde=True, color='teal')
    plt.title("Distribution of Document Lengths (in words)")
    plt.xlabel("Number of Words per Document")
    plt.ylabel("Frequency")
    plt.show()

    # --- 4. POS Tag Frequency ---
    all_pos = [tag for _, tag in pos_tag(all_tokens)]
    pos_counts = Counter(all_pos).most_common()
    pos_tags, pos_freqs = zip(*pos_counts)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(pos_tags), y=list(pos_freqs), palette="coolwarm")
    plt.title("Part of Speech (POS) Tag Frequency")
    plt.xlabel("POS Tag")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

    # --- 5. Bigram Frequency Plot ---
    bigrams = list(ngrams(all_tokens, 2))
    bigram_freq = Counter(bigrams).most_common(top_n_words)
    bigram_labels = [' '.join(b) for b, _ in bigram_freq]
    bigram_counts = [count for _, count in bigram_freq]

    plt.figure(figsize=(10, 5))
    sns.barplot(x=bigram_counts, y=bigram_labels, palette="magma")
    plt.title(f"Top {top_n_words} Bigrams")
    plt.xlabel("Count")
    plt.ylabel("Bigram")
    plt.show()
```

```python
def sbert_feature_extraction(texts, model_name='sentence-transformers/all-MiniLM-L6-v2', batch_size=32):
    """Extract SBERT features from texts with tqdm progress bar."""
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    model.eval()

    all_embeddings = []

    for i in tqdm(range(0, len(texts), batch_size), desc="Extracting SBERT embeddings"):
        batch_texts = texts[i:i+batch_size]
        encoded_input = tokenizer(batch_texts, padding=True, truncation=True, return_tensors='pt')

        with torch.no_grad():
            model_output = model(**encoded_input)

        token_embeddings = model_output.last_hidden_state
        attention_mask = encoded_input['attention_mask']
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()

        sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, dim=1)
        sum_mask = input_mask_expanded.sum(dim=1)
        embeddings = sum_embeddings / sum_mask

        all_embeddings.append(embeddings.cpu().numpy())

    X = np.vstack(all_embeddings)
    return X, model
```

```python
def feature_extraction(X_text, method='SBERT_optimized_QA', model_dir='model', is_train=True):

    x_file = os.path.join(model_dir, f'X_{method}.pkl')

    print(f"[INFO] Feature extraction method: {method}, Train mode: {is_train}")
    model_name = 'sentence-transformers/multi-qa-MiniLM-L6-cos-v1'

    if is_train:
        if os.path.exists(x_file):
            print(f"[INFO] Loading cached Lightweight RoBERTa features from {x_file}")
            X = joblib.load(x_file)
        else:
            print("[INFO] Computing SBERT_semantic_similarity features...")
            X, model = sbert_feature_extraction(X_text, model_name=model_name)
            joblib.dump(X, x_file)
    else:
        print("[INFO] Performing SBERT_semantic_similarity feature extraction for testing...")
        X, model = sbert_feature_extraction(X_text, model_name=model_name)

    return X
```

```python
def train_multioutput_sgd_classifier(Algorithm_prefix, features_dict, labels_dict):
    sgd_models = {}

    for target_name, y_encoded in labels_dict.items():
        features = features_dict.get(target_name)  # 🔥 FIXED: Match keys directly

        if features is None:
            print(f"[ERROR] No features found for target: {target_name}")
            continue

        model_path = f"model/{Algorithm_prefix}_{target_name}_sgd_model.pkl"
        algo_name = f"{Algorithm_prefix} SGD Classifier [{target_name}]"

        X_train, X_test, y_train, y_test = train_test_split(features, y_encoded, test_size=0.2, random_state=42)

        if os.path.exists(model_path):
            print(f"Loading existing SGDClassifier model for {target_name}...")
            sgd = joblib.load(model_path)
        else:
            print(f"Training SGDClassifier for target: {target_name}...")
            sgd = SGDClassifier(loss='log_loss', max_iter=1000, tol=1e-3, random_state=42)
            sgd.fit(X_train, y_train)
            joblib.dump(sgd, model_path)

        y_pred = sgd.predict(X_test)

        # Try getting probabilities (only supported if loss='log_loss')
        try:
            y_score = sgd.predict_proba(X_test)
        except AttributeError:
            try:
                y_score = expit(sgd.decision_function(X_test).reshape(-1, 1))
            except:
                y_score = None

        # Use target-specific metrics_calculator
        if target_name == "relevant":
            metrics_calculator_wlb.calculate_metrics(algo_name, y_pred, y_test, y_score)
        elif target_name == "type":
            metrics_calculator_cg.calculate_metrics(algo_name, y_pred, y_test, y_score)
        else:
            print(f"No metrics_calculator defined for: {target_name}")

        sgd_models[target_name] = sgd

    return sgd_models
```

```python
from sklearn.naive_bayes import ComplementNB
def train_multioutput_complementnb_classifier(Algorithm_prefix, features_dict, labels_dict):
    cnb_models = {}

    for target_name, y_encoded in labels_dict.items():
        features = features_dict.get(target_name)

        if features is None:
            print(f"[ERROR] No features found for target: {target_name}")
            continue

        model_path = f"model/{Algorithm_prefix}_{target_name}_cnb_model.pkl"
        algo_name = f"{Algorithm_prefix} ComplementNB Classifier [{target_name}]"

        X_train, X_test, y_train, y_test = train_test_split(features, y_encoded, test_size=0.2, random_state=42)

        if os.path.exists(model_path):
            print(f"Loading existing ComplementNB model for {target_name}...")
            cnb = joblib.load(model_path)
        else:
            print(f"Training ComplementNB for target: {target_name}...")
            cnb = ComplementNB()
            cnb.fit(X_train, y_train)
            joblib.dump(cnb, model_path)

        y_pred = cnb.predict(X_test)

        # Get predicted probabilities if available
        try:
            y_score = cnb.predict_proba(X_test)
        except:
            try:
                y_score = expit(cnb.decision_function(X_test).reshape(-1, 1))
            except:
                y_score = None

        # Use target-specific metrics calculator
        if target_name == "relevant":
            metrics_calculator_wlb.calculate_metrics(algo_name, y_pred, y_test, y_score)
        elif target_name == "type":
            metrics_calculator_cg.calculate_metrics(algo_name, y_pred, y_test, y_score)
        else:
            print(f"No metrics_calculator defined for: {target_name}")

        cnb_models[target_name] = cnb

    return cnb_models
```

```python
def plot_training_history(history, title=None):
    """Plot training and validation accuracy/loss."""
    if history is None or not isinstance(history, dict):
        print("[ERROR] Invalid history format.")
        return

    plt.figure(figsize=(12, 5))

    # Accuracy Plot
    if 'accuracy' in history:
        plt.subplot(1, 2, 1)
        plt.plot(history['accuracy'], label='Train Acc')
        if 'val_accuracy' in history:
            plt.plot(history['val_accuracy'], label='Val Acc')
        plt.title(f"{title} Accuracy" if title else "Accuracy")
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()

    # Loss Plot
    if 'loss' in history:
        plt.subplot(1, 2, 2)
        plt.plot(history['loss'], label='Train Loss')
        if 'val_loss' in history:
            plt.plot(history['val_loss'], label='Val Loss')
        plt.title(f"{title} Loss" if title else "Loss")
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

    plt.tight_layout()
    plt.show()
def plot_all_histories_from_folder(folder_path="model"):
    """Automatically read and plot all *_history.npy files in a folder."""
    history_files = [f for f in os.listdir(folder_path) if f.endswith("_history.npy")]

    if  history_files:
        print(history_files)

    for file in history_files:
        full_path = os.path.join(folder_path, file)
        try:
            history = np.load(full_path, allow_pickle=True).item()
            plot_training_history(history, title=file.replace("_history.npy", ""))
        except Exception as e:
            print(f"[ERROR] Could not load {file}: {e}")
```

```python
def train_dense_sbert_classifier_extract_features(features_dict, labels_dict, model_prefix="SBERT"):
    extracted_features_dict = {}

    for target_name, y in labels_dict.items():
        X = features_dict[target_name]
        
        # File paths for each target
        model_path = f"model/{model_prefix}_{target_name}_dense_model.h5"
        feature_path = f"model/{model_prefix}_{target_name}_features.npy"
        history_path = f"model/{model_prefix}_{target_name}_history.npy"

        if all(os.path.exists(p) for p in [model_path, feature_path, history_path]):
            print(f"✅ [LOAD] {target_name}: Model and features found. Loading...")
            model = load_model(model_path)
            features = np.load(feature_path)
            extracted_features_dict[target_name] = features
            continue

        print(f"🚀 [TRAIN] {target_name}: No saved model found. Training new Dense model...")

        num_classes = len(np.unique(y))
        y_categorical = to_categorical(y, num_classes=num_classes)

        input_layer = Input(shape=(X.shape[1],))
        x = Dense(512, activation='relu')(input_layer)
        x = Dense(256, activation='relu')(x)
        x = Dense(128, activation='relu', name='feature_layer')(x)
        x = Dense(64, activation='relu')(x)
        output_layer = Dense(num_classes, activation='softmax')(x)

        model = Model(inputs=input_layer, outputs=output_layer)
        model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

        history = model.fit(
            X, y_categorical,
            epochs=50,
            batch_size=32,
            validation_split=0.2,
            verbose=1
        )

        # Extract features from mid-layer
        feature_extractor = Model(inputs=model.input, outputs=model.get_layer('feature_layer').output)
        features = feature_extractor.predict(X)

        # Save results
        model.save(model_path)
        np.save(feature_path, features)
        np.save(history_path, history.history)

        print(f"💾 [SAVED] {target_name}: Model, features, and training history saved.")
        extracted_features_dict[target_name] = features

    return extracted_features_dict
```

```python
def train_multioutput_lda(Algorithm_prefix, features_dict, labels_dict):
    """
    Train and evaluate LDA models for each target in a multi-output classification task.
    """
    model_results = {}

    for target_name, y_encoded in labels_dict.items():
        features = features_dict[target_name]
        model_path = f"model/{Algorithm_prefix}_{target_name}_LDA_model.pkl"
        algo_name = f"{Algorithm_prefix} LDA [{target_name}]"

        X_train, X_test, y_train, y_test = train_test_split(
            features, y_encoded, test_size=0.2, random_state=42
        )

        if os.path.exists(model_path):
            print(f"Loading existing LDA model for {target_name}...")
            model = joblib.load(model_path)
        else:
            print(f"Training LDA for target: {target_name}...")
            model = LDA()
            model.fit(X_train, y_train)
            joblib.dump(model, model_path)

        y_pred = model.predict(X_test)

        try:
            y_score = model.predict_proba(X_test)
        except:
            y_score = None

        # Metrics calculation (replace with your custom functions)
        if target_name == "relevant":
            metrics_calculator_wlb.calculate_metrics(algo_name, y_pred, y_test, y_score)
        elif target_name == "type":
            metrics_calculator_cg.calculate_metrics(algo_name, y_pred, y_test, y_score)
        else:
            print(f"No metrics_calculator defined for: {target_name}")

        model_results[target_name] = model

    return model_results
```

```python
def test_DNN_feature_extraction(features_test, model_base_name="SBERT OQA DNN", model_dir="model"):
    feature_outputs = {}
    targets = ["type", "relevant"]

    for target in targets:
        model_filename = f"{model_base_name}_{target}_dense_model.h5"
        model_path = os.path.join(model_dir, model_filename)

        if not os.path.exists(model_path):
            print(f"❌ Model not found: {model_path}")
            continue

        print(f"✅ Loading model for {target}: {model_path}")
        model = load_model(model_path)

        try:
            feature_extractor = Model(inputs=model.input,
                                      outputs=model.get_layer("feature_layer").output)
        except:
            raise ValueError(f"Model does not contain a layer named 'feature_layer': {model_path}")

        print(f"🚀 Extracting features for {target}...")
        extracted_features = feature_extractor.predict(features_test)
        feature_outputs[target] = extracted_features

    return feature_outputs
```

```python
features_test = feature_extraction(df_test,method='SBERT_optimized_QA',is_train=None)
feature_outputs_dict = test_DNN_feature_extraction(features_test)
```

```python
full_type_classes = ['botnet', 'ddos', 'general', 'ransomware', 'vulnerability', 'leak', 'Zero Day', 'All']
full_relevant_classes = ['False', 'True']

# Loop over each target
for target in ['type', 'relevant']:
    # Get extracted features for current target
    target_features = feature_outputs_dict[target]

    # Predict using the model
    y_pred = LDA_models[target].predict(target_features)

    # Fix label encoder if needed
    if target == 'type':
        label_encoders[target].classes_ = np.array(full_type_classes)
    elif target == 'relevant':
        label_encoders[target].classes_ = np.array(full_relevant_classes)

    # Inverse transform to get original string labels
    mapped_labels = label_encoders[target].inverse_transform(y_pred)

    # Add prediction column
    df_result[f'Predicted_{target}'] = mapped_labels

# Final output
df_result
```
